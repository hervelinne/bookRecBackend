# Functional describe dataset

```python
def describe(df):
    """Return a Pandas dataframe describing the contents of a source dataframe including missing values."""
    
    variables = []
    dtypes = []
    count = []
    unique = []
    missing = []
    
    
    for item in df.columns:
        variables.append(item)
        dtypes.append(df[item].dtype)
        count.append(len(df[item]))
        unique.append(len(df[item].unique()))
        missing.append(df[item].isna().sum())
        

    output = pd.DataFrame({
        'variable': variables, 
        'dtype': dtypes,
        'count': count,
        'unique': unique,
        'missing value': missing
    })    
        
    return output
```

# Content Based Recommendation Engine - Book Title (only)

```python
# logic for recommending books based on Book title. It takes book title and genre as an input.

def recommend(title, genre):
    # Matching the genre with the dataset and reset the index
    temp = df[(df['genre1'] == genre) | (df['genre2'] == genre) | (df['genre3'] == genre)]  
    temp.reset_index(level = 0, inplace = True) 
  
    # Convert the index into series
    indices = pd.Series(temp.index, index = temp['book'])
    
    #Converting the book title into vectors and used bigram
    tf = TfidfVectorizer(analyzer='word', ngram_range=(2, 2), min_df = 1, stop_words='english')
    tfidf_matrix = tf.fit_transform(temp['book'])
    
    # Calculating the similarity measures based on Cosine Similarity
    sg = cosine_similarity(tfidf_matrix, tfidf_matrix)
    
    # Get the index corresponding to original_title
    idx = indices[title]
    
    # Get the pairwsie similarity scores 
    sig = list(enumerate(sg[idx]))
    
    # Sort the books
    sig = sorted(sig, key=lambda x: x[1], reverse=True)
    
    # Scores of the 5 most similar books 
    sig = sig[1:6]
    
    # Book indicies
    book_indices = [i[0] for i in sig]
   
    # Top 5 book recommendation
    rec = temp[['book']].iloc[book_indices]
    print(f"People who read \'{title}\' might also like: {', '.join(i for i in rec['book']) }")
```

```python
# try
recommend("The Trojan Women", "Plays")
```

```python
# Function to calculate precision
def calculate_precision(actual_ratings, recommended_books):
    true_positives = set(actual_ratings[actual_ratings == 2].index) & set(recommended_books.index)
    all_recommended = set(recommended_books.index)
    
    if not all_recommended:
        return 0.0
    
    precision = len(true_positives) / len(all_recommended)
    return precision

# Function to calculate recall
def calculate_recall(actual_ratings, recommended_books):
    true_positives = set(actual_ratings[actual_ratings == 2].index) & set(recommended_books.index)
    all_actual = set(actual_ratings[actual_ratings == 2].index)
    
    if not all_actual:
        return 0.0
    
    recall = len(true_positives) / len(all_actual)
    return recall

# Function to calculate F1-score
def calculate_f1_score(precision, recall):
    if precision + recall == 0:
        return 0.0
    f1_score = 2 * (precision * recall) / (precision + recall)
    return f1_score

# Aggregate scores for the entire model
precision_scores = []
recall_scores = []
f1_scores = []

for user_id in test_data['user_id'].unique():
    user_ratings = test_data[test_data['user_id'] == user_id]['rating']
    recommendations_knn = get_book_recommendations_knn(user_id)  # Use your KNN-based recommendation function
    
    precision = calculate_precision(user_ratings, recommendations_knn)
    recall = calculate_recall(user_ratings, recommendations_knn)
    f1_score = calculate_f1_score(precision, recall)
    
    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1_score)

# Calculate average scores for the entire model
average_precision = sum(precision_scores) / len(precision_scores)
average_recall = sum(recall_scores) / len(recall_scores)
average_f1_score = sum(f1_scores) / len(f1_scores)

print(f"Average Precision: {average_precision}")
print(f"Average Recall: {average_recall}")
print(f"Average F1-Score: {average_f1_score}")
```

```python
# Encode user and book IDs
user_encoder = LabelEncoder()
book_encoder = LabelEncoder()

df['user_id'] = user_encoder.fit_transform(df['user_id'])
df['book_id'] = book_encoder.fit_transform(df['book_id'])

# Define the neural network model
def collaborative_filtering_model(num_users, num_books, embedding_size=50):
    # User embedding layer
    user_input = Input(shape=(1,), name='user_input')
    user_embedding = Embedding(input_dim=num_users, output_dim=embedding_size)(user_input)
    user_embedding = Flatten()(user_embedding)

    # Book embedding layer
    book_input = Input(shape=(1,), name='book_input')
    book_embedding = Embedding(input_dim=num_books, output_dim=embedding_size)(book_input)
    book_embedding = Flatten()(book_embedding)

    # Dot product of user and book embeddings
    dot_product = Dot(axes=1)([user_embedding, book_embedding])

    # Combine user and book IDs
    merged = Concatenate()([user_embedding, book_embedding])

    # Fully connected layers
    dense1 = tf.keras.layers.Dense(128, activation='relu')(merged)
    dense2 = tf.keras.layers.Dense(64, activation='relu')(dense1)

    # Output layer
    output = tf.keras.layers.Dense(1)(dense2)

    model = Model(inputs=[user_input, book_input], outputs=output)
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    return model

# Create and compile the model
num_users = df['user_id'].nunique()
num_books = df['book_id'].nunique()
model = collaborative_filtering_model(num_users, num_books)

# Prepare data for training
train_user_ids = train_data['user_id'].values
train_book_ids = train_data['book_id'].values
train_ratings = train_data['rating'].values

# Train the model
history = model.fit([train_user_ids, train_book_ids], train_ratings, epochs=10, batch_size=64, validation_split=0.2)

# Evaluate on the test set
test_user_ids = test_data['user_id'].values
test_book_ids = test_data['book_id'].values
test_ratings = test_data['rating'].values

predictions = model.predict([test_user_ids, test_book_ids])
mse = mean_squared_error(test_ratings, predictions)

print(f"Mean Squared Error on Test Set: {mse}")
```
